## Duplicates
Check if there are more than six rows stored the same value consecutively within column Class and Speed

```python
df1["SpeedClass"] = df1["Speed"].astype(str) + "-" + df1["Class"].astype(str)
df1.head()
```

```python
def duplicatedSpeedClass(x):
    #define length variable
    length = len(x)
    #define duplicates list
    duplicates = []
    
    #use for loop within range from first row to last row
    for i in range(0, length-1):
        #this conditional statement aims to find five consecutive duplicacies, should you need to find more or less 
        #than five duplicacies, edit the following statement by adding or removing "and x[i] == x[i+(n)"]
        if x[i] == x[i+1] and x[i] == x[i+2] and x[i] == x[i+3] and x[i] == x[i+4]:
                #add the index into "duplicates" list. If the conditional statement above was changed, the list 
                #inside extend() method should follow accordingly
                duplicates.extend([i, i+1, i+2, i+3, i+4])
                #sometimes there will be duplicacies within "duplicates" list above. To remove any duplicacies, we need to
                #convert the "duplicates" list to a set and then back to a list, cause sets only contain unique elements
                duplicates_unique = list(set(duplicates))
    
    #the "duplicates_uniques" is the final result in this function
    return duplicates_unique

print(duplicatedSpeedClass(df1["SpeedClass"]))
```

```python
df1.iloc[duplicatedSpeedClass(df1["SpeedClass"])]
```
