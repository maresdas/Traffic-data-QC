## Additional QC

### 1. Speed greater than 85 mph

```python
#the values above 85 is fileterd from "Speed" column. Then it is sorted from low to high
df1[df1["Speed"] > 85].sort_values("Speed")
```

```python
#the speed above 85 is then grouped (.groupby) and counted (.count)
df1[df1["Speed"] > 85].groupby('Speed').Speed.count()
```

### 2. Speed less than 10 mph

```python
#the values below 10 is fileterd from "Speed" column. Then it is sorted from low to high
df1[df1["Speed"] < 10].sort_values("Speed")
```

```python
#the speed below 10 is then grouped (.groupby) and counted (.count)
df1[df1["Speed"] < 10].groupby('Speed').Speed.count()
```

### 3. Vehicle with GVW greater than 250 kips

```python
#the values above 250 is fileterd from "GVW" column. Then it is sorted from low to high
df1[df1["GVW"] > 250].sort_values("GVW")
```

```python
#the speed above 250 is then grouped (.groupby) and counted (.count)
df1[df1["GVW"] > 250].groupby("GVW").GVW.count()
```

### 4. Class 1 with GVW more than 3.0 kips

```python
#Criteria 1 (Cr1): Class 1
Cr1 = df1["Class"] == 1

#Criteria 2 (Cr2): GVW value above 3
Cr2 = df1["GVW"] > 3

#Combine Cr1 and Cr2
df1[Cr1 & Cr2]
```

```python
#the class 1 with GVW above 3 kips is then grouped (.groupby) and counted (.count)
df1[Cr1 & Cr2].groupby("GVW").GVW.count()
```

### 5. Class 2 & 3 with GVW more than 15 kips

```python
#Criteria 3 (Cr3): Class 2 & 3
Cr3 = df1["Class"].between(2, 3)

#Criteria 4 (Cr4): GVW value above 15
Cr4 = df1["GVW"] > 15

#Combine Cr3 and Cr4
df1[Cr3 & Cr4]
```

```python
#the class 2 & 3 with GVW above 15 kips is then grouped (.groupby) and counted (.count)
df1[Cr3 & Cr4].groupby("GVW").GVW.count()
```

### 6. ESALs greater than 70

```python
#the values above 70 is fileterd from "Esal" column. Then it is sorted from low to high
df1[df1["ESAL"] > 70].sort_values("ESAL")
```

```python
#it is then grouped (.groupby) and counted (.count)
df1[df1["ESAL"] > 70].groupby("ESAL").count()
```

### 7. Wt1 greater than 25 kips

```python
#the values above 25 is fileterd from "Weight A" column. Then it is sorted from low to high
df1[df1["Weight A"] > 25].sort_values("Weight A")
```

```python
#it is then grouped (.groupby) and counted (.count)
df1[df1["Weight A"] > 25].groupby("Weight A").count()
```
